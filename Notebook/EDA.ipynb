```python
def plot_nan_percentage(df):
    """
    Plot the percentage of NaN values in each column of the DataFrame.

    Parameters:
    - df: pandas DataFrame

    Returns:
    - None (plots the figure)
    """
    nan_percentage = (df.isnull().mean() * 100).round(2)
    nan_percentage = nan_percentage[nan_percentage > 0]  # Exclude columns with no NaN values

    plt.figure(figsize=(10, 6))
    nan_percentage.plot(kind='bar', color='skyblue')
    plt.title('Percentage of NaN Values in Each Column')
    plt.xlabel('Columns')
    plt.ylabel('Percentage of NaN Values')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

plot_nan_percentage(df=df)
```

```python
def separate_discrete_continuous_columns(df, threshold=10):
    """
    Separate discrete and continuous columns in the DataFrame.

    Parameters:
    - df: pandas DataFrame
    - threshold: Maximum number of unique values to consider a column as discrete (default is 10)

    Returns:
    - discrete_columns: list of discrete column names
    - continuous_columns: list of continuous column names
    """
    discrete_columns = []
    continuous_columns = []

    for column in df.columns:
        if df[column].dtype == 'object' or df[column].nunique() <= threshold:
            discrete_columns.append(column)
        else:
            continuous_columns.append(column)

    return discrete_columns, continuous_columns


discrete_columns, continuous_columns=separate_discrete_continuous_columns(df[numerical_columns], threshold=10)
```

```python
def change_to_number(freight_cost_usd):
    regex = {"id_number": ":\d*"}
    match = re.search(regex['id_number'], freight_cost_usd, re.IGNORECASE)
    
    if match:
        id = match.group(0).replace(':', '')
        filtered = df.query("ID == " + id)
        
        if not filtered.empty:
            return filtered['Freight_Cost_(USD)'].iloc[0]
    
    return freight_cost_usd

def convert_to_number(weight):
    regex = {"id_number": ":\d*"}
    match = re.search(regex['id_number'], weight, re.IGNORECASE)
    
    if match:
        id = match.group(0).replace(':', '')
        filtered = df.query("ID == " + id)
        
        if not filtered.empty:
            return filtered['Weight_(Kilograms)'].iloc[0]
    
    return weight

# Apply the functions to clean columns
df['Freight_Cost_USD_Clean'] = df['Freight_Cost_(USD)'].apply(change_to_number)
df['Weight_Kilograms_Clean'] = df['Weight_(Kilograms)'].apply(convert_to_number)

print("Weight and freight completed")

# Identify indexes with specific conditions
freight_cost_indexes = df.index[(df['Freight_Cost_USD_Clean'] == 'Freight Included in Commodity Cost') | (df['Freight_Cost_USD_Clean'] == 'Invoiced Separately')].tolist()
weight_indexes = df.index[df['Weight_Kilograms_Clean'] == 'Weight Captured Separately'].tolist()
shipment_indexes = df.index[df['Shipment_Mode'] == 'no_value'].tolist()

print("Freight_Cost_USD_Clean_indexes:", len(freight_cost_indexes))
print("Weight_Kilograms_Clean_indexes:", len(weight_indexes))
print("Shipment_Mode indexes:         ", len(shipment_indexes))

# Combine indexes and drop rows
indexes = list(set(freight_cost_indexes + weight_indexes + shipment_indexes))
print("Indexes:", len(indexes))
df_clean = df.drop(indexes)
```

```python
def histoplot_dataframe(df, bins=50, color='skyblue'):
    """
    Generate histogram plots for each column in the DataFrame in a single row.

    Parameters:
    - df: pandas DataFrame
    - bins: int or sequence, optional (default is 10)
    - color: str, optional (default is 'skyblue')

    Returns:
    - None (displays the plots in a single row)
    """
    num_columns = len(df.columns)
    num_plots_per_row = 3  # Number of plots to display in each row
    num_rows = -(-num_columns // num_plots_per_row)  # Ceiling division to calculate the number of rows

    fig, axes = plt.subplots(num_rows, num_plots_per_row, figsize=(15, 5 * num_rows))

    for i, column in enumerate(df.columns):
        row_index = i // num_plots_per_row
        col_index = i % num_plots_per_row

        ax = axes[row_index, col_index] if num_rows > 1 else axes[col_index]
        ax.hist(df[column], bins=bins, color=color, edgecolor='black')
        ax.set_title(f'Histogram Plot for {column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Frequency')
        ax.grid(axis='y', alpha=0.75)

    plt.tight_layout()
    plt.show()
histoplot_dataframe(df=df[numerical_columns])
```

```python
def replace_nan_with_median(df, column_label):
    """
    Replace NaN values in a specific column with the median.

    Parameters:
    - df: pandas DataFrame
    - column_label: str, column label for which NaN values should be replaced

    Returns:
    - df: pandas DataFrame with NaN values replaced by median in the specified column
    """
    median_value = df[column_label].median()
    df[column_label] = df[column_label].fillna(median_value)
    return df

df=replace_nan_with_median(df,column_label="Line_Item_Insurance_(USD)")
```

```python
def detect_and_plot_outliers_histogram(df, column_list, quantile_percentage=0.98):
    """
    Detect and plot outliers in specified columns of a DataFrame using histograms.

    Parameters:
    - df: pandas DataFrame
    - column_list: list of column names for which outliers should be detected and plotted
    - quantile_percentage: float, the quantile percentage to determine the lower and upper bounds for outliers (default is 0.75)

    Returns:
    - None (displays histograms with outliers highlighted for each specified column)
    """
    num_columns = len(column_list)
    num_rows = (num_columns + 2) // 3  # Number of rows needed for subplots
    num_cols = min(num_columns, 3)  # Number of columns for subplots

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))
    axes = axes.flatten()  # Flatten the 2D array of subplots

    for i, column in enumerate(column_list):
        ax = axes[i]

        # Calculate the lower and upper quantiles
        Q1 = df[column].quantile(1 - quantile_percentage)
        Q3 = df[column].quantile(quantile_percentage)

        # Calculate the interquartile range (IQR)
        IQR = Q3 - Q1

        # Define the lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Detect outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        # Plot the histogram with outliers highlighted
        ax.hist(df[column], bins=20, color='skyblue', edgecolor='black', label='Data Distribution')
        ax.hist(outliers[column], bins=5, color='red', edgecolor='black', label='Outliers')
        ax.set_title(f'{column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Frequency')
        ax.legend()

    # Adjust layout to prevent overlapping
    plt.tight_layout()
    plt.show()


detect_and_plot_outliers_histogram(df, column_list=numerical_columns)
```

```python
def remove_outliers_and_plot(df, column_list, quantile_percentage=0.99):
    """
    Remove outliers from specified columns in a DataFrame using a specified method and
    plot box plots before and after the removal side by side.

    Parameters:
    - df: pandas DataFrame
    - column_list: list of column names for which outliers should be removed
    - quantile_percentage: float, quantile percentage for outlier removal (default is 0.9)

    Returns:
    - df_clean: pandas DataFrame with outliers removed
    """
    num_columns = len(column_list)

    # Print shape before removal
    print(f"Shape before outlier removal: {df.shape}")

    # Create subplots for box plots
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plot box plots before outlier removal
    for i, column in enumerate(column_list):
        axes[0].boxplot(df[column], positions=[i + 1], widths=0.6, labels=[column])
    axes[0].set_title('Box Plots Before Outlier Removal')
    axes[0].set_ylabel('Values')
    axes[0].tick_params(axis='x', rotation=90)  # Rotate x-labels by 90 degrees

    # Remove outliers based on the specified quantile percentage
    df_clean = df.copy()
    for column in column_list:
        upper_bound = df_clean[column].quantile(quantile_percentage)
        df_clean = df_clean[df_clean[column] <= upper_bound]

    # Print shape after removal
    print(f"Shape after outlier removal: {df_clean.shape}")

    # Plot box plots after outlier removal
    for i, column in enumerate(column_list):
        axes[1].boxplot(df_clean[column], positions=[i + 1], widths=0.6, labels=[column])
    axes[1].set_title('Box Plots After Outlier Removal')
    axes[1].set_ylabel('Values')
    axes[1].tick_params(axis='x', rotation=90)

    plt.tight_layout()
    plt.show()

    return df_clean


df=remove_outliers_and_plot(df, column_list=outlier_removal)
```

```python
def replace_low_frequency_categories(dataframe, column_name, threshold_percentage=1, replace_value='Others'):
    """
    Replace categories with less than a specified percentage occurrence in a column with a specified replacement value.

    Parameters:
    - dataframe: pandas DataFrame
    - column_name: str, the name of the column with categorical data
    - threshold_percentage: float, the threshold percentage for category occurrence (default is 5)
    - replace_value: str, the replacement value for low-frequency categories (default is 'Others')

    Returns:
    - None (modifies the DataFrame in place)
    """
    # Calculate the percentage of each category
    category_percentages = dataframe[column_name].value_counts(normalize=True) * 100

    # Identify categories with less than the threshold percentage
    low_frequency_categories = category_percentages[category_percentages < threshold_percentage].index.tolist()

    # Replace low-frequency categories with the specified replacement value
    dataframe[column_name] = np.where(dataframe[column_name].isin(low_frequency_categories), replace_value, dataframe[column_name])
```

```python
def plot_categorical_hue(dataframe, x_column, hue_column):
    """
    Plot a categorical column with unique values of another column as hue.

    Parameters:
    - dataframe: pandas DataFrame
    - x_column: str, the name of the column to be plotted on the x-axis
    - hue_column: str, the name of the column to be used as hue

    Returns:
    - None (displays the plot)
    """
    plt.figure(figsize=(12, 8))
    sns.countplot(x=x_column, hue=hue_column, data=dataframe, palette='viridis')
    plt.title(f'{x_column} with {hue_column} as Hue')
    plt.xlabel(x_column)
    plt.ylabel('Count')
    plt.show()
```

```python
def plot_stacked_bar(df, x_column, hue_column):
    """
    Plot a stacked bar graph for two categorical columns.

    Parameters:
    - df: pandas DataFrame
    - x_column: str, the name of the column for the x-axis
    - hue_column: str, the name of the column to be used for stacking

    Returns:
    - None (displays the stacked bar graph)
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(x=x_column, hue=hue_column, data=df, palette='viridis')
    plt.title(f'Stacked Bar Graph - {x_column} vs {hue_column}')
    plt.xlabel(x_column)
    plt.ylabel('Count')
    plt.legend(title=hue_column)
    plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_country_frequency_on_map(dataframe, country_column):
    """
    Plot countries on a world map with color-coding based on frequency count.

    Parameters:
    - dataframe: pandas DataFrame with country labels
    - country_column: str, the name of the column containing country labels

    Returns:
    - None (displays the plot)
    """
    # Count the frequency of each country
    country_counts = dataframe[country_column].value_counts().reset_index()
    country_counts.columns = ['name', 'frequency']

    # Load the world map shapefile
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

    # Merge the world map with the DataFrame on the specified column
    merged_data = world.merge(country_counts, how='left', on='name')

    # Plot the world map with country color-coding based on frequency count
    fig, ax = plt.subplots(figsize=(15, 10))
    merged_data.boundary.plot(ax=ax, linewidth=0.8)
    merged_data.plot(column='frequency', ax=ax, legend=True,
                    legend_kwds={'label': 'Frequency Scale', 'orientation': 'horizontal'})
    plt.title(f'Country Color Coding on World Map based on Frequency')
    plt.show()

plot_country_frequency_on_map(dataframe=df,country_column='Country')
```

```python
import pandas as pd
import folium

def plot_world_map_with_shipment_modes(df, location_column='Location', shipment_mode_column='Shipment_Mode', threshold=3):
    """
    Plot a world map with markers color-coded based on the count frequency of shipment modes.

    Parameters:
    - df: pandas DataFrame
    - location_column: str, the column containing location names or labels
    - shipment_mode_column: str, the column containing shipment mode values
    - threshold: int, count threshold for color-coding markers (default is 3)

    Returns:
    - None (displays the interactive map)
    """
    # Calculate the frequency of each shipment mode
    shipment_mode_counts = df[shipment_mode_column].value_counts()

    # Create a folium map centered around the world
    world_map = folium.Map(location=[0, 0], zoom_start=2)

    # Iterate through DataFrame rows and add markers to the map
    for index, row in df.iterrows():
        shipment_mode_count = shipment_mode_counts.get(row[shipment_mode_column], 0)

        # Customize colors based on count threshold
        color = 'green' if shipment_mode_count >= threshold else 'blue'

        folium.Marker(
            location=[0, 0],  # Replace with actual latitude and longitude values for each location
            popup=f"{location_column}: {row[location_column]}, {shipment_mode_column}: {row[shipment_mode_column]}, Count: {shipment_mode_count}",
            icon=folium.Icon(color=color)
        ).add_to(world_map)

    # Display the map
    display(world_map)


```

```python
def histoplot_dataframe_column(dataframe, column_label):
    """
    Plot a histogram for a specified column in a DataFrame.

    Parameters:
    - dataframe: pandas DataFrame
    - column_label: str, name of the column for which to plot the histogram

    Returns:
    - None (displays the histogram)
    """
    # Check if the column exists in the DataFrame
    if column_label not in dataframe.columns:
        print(f"Column '{column_label}' not found in the DataFrame.")
        return

    # Create a histogram using seaborn
    plt.figure(figsize=(8, 6))
    sns.histplot(dataframe[column_label], bins=20, kde=True, color='skyblue', edgecolor='black')

    # Customize plot appearance
    plt.title(f'Histogram for {column_label}')
    plt.xlabel(column_label)
    plt.ylabel('Frequency')

    # Display the plot
    plt.show()

histoplot_dataframe_column(dataframe=df,column_label='Freight_Cost_USD_Clean'
                           )
```

```python
def scatterplot_with_hue(dataframe, x_column, y_column, hue_column):
    """
    Create a scatter plot with hue for two numerical columns in a DataFrame.

    Parameters:
    - dataframe: pandas DataFrame
    - x_column: str, name of the x-axis numerical column
    - y_column: str, name of the y-axis numerical column
    - hue_column: str, name of the column to use for hue

    Returns:
    - None (displays the scatter plot)
    """
    # Check if the columns exist in the DataFrame
    if x_column not in dataframe.columns or y_column not in dataframe.columns or hue_column not in dataframe.columns:
        missing_columns = [col for col in [x_column, y_column, hue_column] if col not in dataframe.columns]
        print(f"Columns {', '.join(missing_columns)} not found in the DataFrame.")
        return

    # Create a scatter plot with hue using seaborn
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x=x_column, y=y_column, hue=hue_column, data=dataframe, palette='viridis', legend='full')

    # Customize plot appearance
    plt.title(f'Scatter Plot for {x_column} vs {y_column} with Hue {hue_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.legend(title=hue_column)

    # Display the plot
    plt.show()
```

```python
def save_dataframe_to_csv(dataframe, folder_name, csv_name):
    """
    Save a DataFrame to a CSV file in the specified folder.

    Parameters:
    - dataframe: pandas DataFrame
    - folder_name: str, name of the folder to save the CSV file
    - csv_name: str, name of the CSV file

    Returns:
    - None
    """
    # Check if the folder exists, if not, create it
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    # Construct the full path to the CSV file
    csv_path = os.path.join(folder_name, csv_name)

    # Save the DataFrame to CSV
    dataframe.to_csv(csv_path, index=False)
```
